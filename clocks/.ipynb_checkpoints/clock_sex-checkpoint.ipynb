```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from clock_utils import make_groups

basepath = "/home/ocampolab/Documents/Data/Cheyenne_Blood_Clock/Paper_data/2021-10-07-10:13-data-and-results_no-peak-filt/"

# Get preprocessed data
data = pd.read_csv(basepath + "atac_data_tpm.csv", index_col=0)
labels = pd.read_csv(basepath + "atac_labels.csv", index_col=0, squeeze=True)
data = data.apply(np.log1p)

groups = make_groups(11, labels)
```

```python
metadata = pd.read_csv("/home/ocampolab/Documents/Data/Cheyenne_Blood_Clock/Sample_metadata_complete.csv", sep=";")
metadata["Sample_name"] = metadata["Sample_name"].replace("ATAC_", "", regex=True)
metadata = metadata.loc[metadata["Sample_name"].isin(data.index)]
np.unique((data.index == metadata["Sample_name"]))

men_snames = metadata["Sample_name"][metadata["Sex"] == "M"].to_numpy()
women_snames = metadata["Sample_name"][metadata["Sex"] == "F"].to_numpy()

rnd = np.random.default_rng(seed=0)
com_men_snames = rnd.choice(men_snames, size=len(women_snames), replace=False)
alt_men_snames = rnd.choice(np.setdiff1d(men_snames, com_men_snames), size=len(women_snames), replace=False)

set_males = np.concatenate([com_men_snames, alt_men_snames])
set_mixed = np.concatenate([com_men_snames, women_snames])

data_males = data.loc[set_males]
data_mixed = data.loc[set_mixed]

labels_males = labels.loc[set_males]
labels_mixed = labels.loc[set_mixed]

groups_males = make_groups(11, labels_males)
groups_mixed = make_groups(11, labels_mixed)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import Pipeline

pipes = list()

# Model 1

pipes.append(Pipeline(steps=[("regressor", ElasticNet(normalize=True, max_iter=5000, tol=0.0005))]))

param_grid = dict()
param_grid["regressor__alpha"] = np.logspace(-5, 0, 40)
param_grid["regressor__l1_ratio"] = [0.7, 0.8, 0.9, 0.95, 0.99, 1]
```

```python
from clock_utils import best_within_1sd_fixed_l1_ratio, best_using_confidence, nested_cv, group_nested_cv
# import warnings
# warnings.filterwarnings("ignore")

all_res = list()
all_res.append(group_nested_cv(data_males, labels_males, groups_males, pipes[0], param_grid, opt_choice_fun=True, to_filter=-1, relax=False))
all_res.append(group_nested_cv(data_mixed, labels_mixed, groups_mixed, pipes[0], param_grid, opt_choice_fun=True, to_filter=-1, relax=False))
```
