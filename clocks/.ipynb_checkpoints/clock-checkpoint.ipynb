```python
import sys
sys.path.append("../scripts")
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from clock_utils import make_groups

%load_ext autoreload
%autoreload 2

basepath = "../data/paper_data/"
datapath = basepath + "atac_data_limma.csv"
labelpath = basepath + "atac_labels_limma.csv"

# Get preprocessed data
# data = pd.read_csv(basepath + "atac_data_tpm.csv", index_col=0)
# labels = pd.read_csv(basepath + "atac_labels.csv", index_col=0, squeeze=True)
# data = data.apply(np.log1p)

data = pd.read_csv(datapath, index_col=0)
labels = pd.read_csv(labelpath, index_col=0, squeeze=True)
# Log is already part of the limma method

groups = make_groups(11, labels)
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_regression

pipes = list()

# Model 1

pipes.append(Pipeline(steps=[("regressor", ElasticNet(max_iter=5000, tol=0.0005))]))
# pipes.append(Pipeline(steps=[
#     ("filter", SelectKBest(f_regression, k=5000)),
#     ("regressor", ElasticNet(max_iter=5000, tol=0.0005))]))

param_grid = dict()
param_grid["regressor__alpha"] = np.logspace(-5, 0, 40) #40
param_grid["regressor__l1_ratio"] = [0.7, 0.8, 0.9, 0.95, 0.99, 1]
```

```python
from clock_utils import group_nested_cv

all_res = list()
all_res.append(group_nested_cv(data, labels, groups, pipes[0], param_grid))
```

```python
from sklearn.model_selection import LeavePGroupsOut, GridSearchCV

# Hardcoded settings
n_jobs=10
inner_groups_out = 1
scoring = "neg_median_absolute_error"

# Safety check for data and label shapes
if data.shape[0] != labels.shape[0]:
    raise ValueError("Shapes of data and labels don't match")
    
# Define inner CV
cv_inner = LeavePGroupsOut(inner_groups_out)

# optimize hyperparameters by GridSearchCV
gridCV = GridSearchCV(pipes[0], param_grid, scoring=scoring, n_jobs=n_jobs, cv=cv_inner, verbose=False, refit=True)
gridCV.fit(data, labels, groups=groups)

final_model = dict()
final_model["cv_results"] = gridCV.cv_results_
final_model["best_estimator"] = gridCV.best_estimator_
final_model["best_params"] = gridCV.best_params_
final_model["best_index"] = gridCV.best_index_
```
