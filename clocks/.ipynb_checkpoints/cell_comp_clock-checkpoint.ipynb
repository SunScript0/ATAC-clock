```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from clock_utils import make_groups

basepath = "/home/ocampolab/Documents/Data/Cheyenne_Blood_Clock/Paper_data/2021-10-07-10:13-data-and-results_no-peak-filt/"

# Load data
atac = pd.read_csv(basepath + "atac_data_tpm.csv", index_col=0)
labels = pd.read_csv(basepath + "atac_labels.csv", index_col=0, squeeze=True)
metadata = pd.read_csv("/home/ocampolab/Documents/Data/Cheyenne_Blood_Clock/Sample_metadata_complete.csv", sep=";")
metadata.replace("ATAC_CR", "CR", regex=True, inplace=True)

# Get cell composition and gender
metadata.index = metadata["Sample_name"]
metadata.drop(columns=["Sample_name"], inplace=True)
snames = atac.index.to_list()
snames.remove("CR_118_1") # does not have cell composition data
# cell composition data
data_cells = metadata.loc[snames, ["Monocytes [%]", "Granulocytes [%]", "Lymphocytes [%]", "B Cells [%]", 
                                      "NK Cells [%]", "T Cells [%]", "CD4+ T Cells [%]", "CD8+ T Cells [%]"]]
data_gender = (metadata.loc[snames, "Sex"] == "F") * 1
data_gender.columns = ["F"]
# gender and cell composition data
data_cells_gender = pd.concat([data_gender, data_cells], axis=1)

# gender, cell composition and atac data
data_cells_gender_atac = pd.concat([data_gender, data_cells, atac.loc[snames]], axis=1)

labels = labels.loc[snames]

# Define groups
groups = make_groups(11, labels)
```

```python
from sklearn.linear_model import ElasticNet, LinearRegression
from sklearn.pipeline import Pipeline

pipes = list()

pipes.append(Pipeline(steps=[("regressor", LinearRegression(normalize=True))]))
pipes.append(Pipeline(steps=[("regressor", ElasticNet(normalize=True, max_iter=5000, tol=0.0005))]))

param_grid = dict()
param_grid["regressor__alpha"] = np.logspace(-5, 0, 40)
param_grid["regressor__l1_ratio"] = [0.7, 0.8, 0.9, 0.95, 0.99, 1]
```

```python
from clock_utils import group_nested_cv
# import warnings
# warnings.filterwarnings("ignore")

all_res = list()
# OLS for cells and gender
all_res.append(group_nested_cv(data_cells, labels, groups, pipes[0], dict(), opt_choice_fun=True))
all_res.append(group_nested_cv(data_cells_gender, labels, groups, pipes[0], dict(), opt_choice_fun=True))
# Elastic net for cells and gender
all_res.append(group_nested_cv(data_cells, labels, groups, pipes[1], param_grid, opt_choice_fun=True))
all_res.append(group_nested_cv(data_cells_gender, labels, groups, pipes[1], param_grid, opt_choice_fun=True))
# Elastic net for cells, gender and atac
all_res.append(group_nested_cv(data_cells_gender_atac, labels, groups, pipes[1], param_grid, opt_choice_fun=True))
```
